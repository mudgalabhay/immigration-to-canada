```python
#Create a histogram of 10000 random numbers
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(10000)
plt.hist(x,100)
plt.title(r"Normal distribution with $\mu=0,\sigma=1$")
plt.savefig("matplot for 10000 random numbers")
plt.show()
```
