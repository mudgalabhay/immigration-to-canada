```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
import seaborn as sns
path = "C:\\Users\\admin\\OneDrive\\Desktop\\Datasets\\Immigration to Canada IBM Dataset\\Canada.xlsx"
print("Data read into pandas dataframe")
df_canada = pd.read_excel(path,"Canada by Citizenship",skiprows=range(20),skipfooter=2)
df_canada.head()
```

```python
#Data Preprocessing/Data Cleaning/Data Wrangling
df_canada.drop(['AREA','Type','Coverage','REG','DEV'],axis = 1,inplace=True)
```

```python
#Renaming features
df_canada.rename(columns={'OdName':'Country','AreaName':'Continent', 'RegName':'Region'},inplace=True)
df_canada.head(2)
```

```python
#Addding a Total feature
df_canada['Total'] = df_canada.sum(axis=1)
df_canada.head(2)
```

```python
#Convert Column name to string
df_canada.columns = list(map(str,df_canada.columns))
#[print (type(x)) for x in df_canada.columns.values]
```

```python
years = list(map(str,range(1980,2014)))
years
```

```python
haiti = df_canada.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.index= haiti.index.map(int)
```

```python
haiti.plot(kind="line")
plt.title("Immigration from Haiti")
plt.ylabel("Number of immigrants")
plt.xlabel("Years")
plt.show()
```

```python
#in the above plot we can notice that how immigrants from haiti spiked up from 2010
#Annotate this spike in the plot
haiti.plot(kind="line")
plt.title("Immigration from Haiti")
plt.ylabel("Number of immigrants")
plt.xlabel("Years")
plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

```python
#Area Plot
df_canada.sort_values(['Total'],ascending=False,inplace=True)
```

```python
df_top5 = df_canada.head()
df_top5 = df_top5[years].head()
df_top5.head()
```

```python
#df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',alpha = 0.35,stacked= False,figsize=(30,20))
plt.title("Immigration trend of top 5 countries")
plt.ylabel("Number of immigrants")
plt.xlabel("Years")
plt.show()
```

```python
#Bar charts
df_iceland = df_canada.loc['Iceland',years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar',figsize=(10,6))
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('Icelandic immigrants to Canada from 1980 to 2013')
plt.show()
#Below bargraph shows the impact of financial crises the number of immigrants to Canada started increasing rapidly after 2008
```

```python
#Pie chart
df_continents = df_canada.groupby('Continent',axis=0).sum()
df_continents['Total'].plot(kind='pie')
plt.title("Immigration to Canada by Continent 1980-2013")
plt.show()
```

```python
#Box Plots
#Creating box plot to visualize immigration from Japan to Canada
df_japan = df_canada.loc[['Japan'],years].transpose()
df_japan.plot(kind='box')
plt.title("Box plot of Japnese Immigration from 1980-2013")
plt.ylabel("Number of Immigrants")
plt.show()
```

```python
#Scatter Plot
#Scatter Plot of total annual immigration to canada from 1980-2013
df_tot = pd.DataFrame(df_canada[years].sum(axis=0))
df_tot.index = map(int, df_tot.index)#changing data-type of years to int
df_tot.reset_index(inplace = True)
df_tot.columns=['years','Total']
#df_tot
df_tot.plot(kind='scatter',x='years',y='Total')
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
```

```python
#Creating regression plot using seaborn
ax =  sns.regplot(x='years',y='Total',data=df_tot,color="green",marker="+")
```
